## Quick and dirty code, only used for rendering the different icons

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

tips = sns.load_dataset("tips")

fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")
```

```python
# Distplot plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")
np.random.seed(0)
x = np.random.randn(100)

ax = sns.distplot(x)

plt.savefig('seaborn_dist.svg')
plt.show()
plt.clf()
```

```python
# Joint plot
# -----------------------------------------------------------------------------
sns.set_style("whitegrid")
penguins = sns.load_dataset("penguins")
ax = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", height=0.8, kind="hex")

plt.savefig('seaborn_joint.svg')
plt.show()
plt.clf()
```

```python
# Kde plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")
np.random.seed(1)
X = 4 + np.random.normal(0,1.5,200)

sns.kdeplot(X)

plt.savefig('seaborn_kde.svg')
plt.show()
plt.clf()
```

```python
# pairplot plot
# -----------------------------------------------------------------------------
sns.set_style("whitegrid")

penguins = sns.load_dataset("penguins")
ax = sns.pairplot(penguins,
                x_vars=["bill_length_mm", "bill_depth_mm"],
                y_vars=["bill_length_mm", "bill_depth_mm"],
                height=2, plot_kws=dict(marker="+", linewidth=0.2))
ax.fig.tight_layout()
plt.savefig('seaborn_pairplot.svg')
plt.show()
plt.clf()
```

```python
# rug plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")

sns.kdeplot(X)
sns.rugplot(X, height=.1)

plt.savefig('seaborn_rugplot.svg')
plt.show()
plt.clf()

```

```python
# Bar plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")

ax = sns.barplot(x="day", y="total_bill", data=tips)

plt.savefig('seaborn_bar.svg')
plt.show()
plt.clf()
```

```python
# Count plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")

ax = sns.countplot(x="day", data=tips)

plt.savefig('seaborn_count.svg')
plt.show()
plt.clf()
```

```python
# Box plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")

ax = sns.boxplot(x="day", y="total_bill", data=tips)

plt.savefig('seaborn_boxplot.svg')
plt.show()
plt.clf()
```

```python
# violin plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")

ax = sns.violinplot(x="day", y="total_bill", data=tips)

plt.savefig('seaborn_violinplot.svg')
plt.show()
plt.clf()
```

```python
# Strip plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")

ax = sns.stripplot(x="day", y="total_bill", data=tips)

plt.savefig('seaborn_stripplot.svg')
plt.show()
plt.clf()
```

```python
# swarm plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")

ax = sns.swarmplot(x="day", y="total_bill", data=tips)

plt.savefig('seaborn_swarmplot.svg')
plt.show()
plt.clf()
```

```python
# heatmap plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")
tips_mx = tips.corr()
ax = sns.heatmap(tips_mx)

plt.savefig('seaborn_heatmap.svg')
plt.show()
plt.clf()
```

```python
# clustermap plot
# -----------------------------------------------------------------------------
fig = plt.figure(figsize=(0.8,0.8))
d = 0.01
ax = fig.add_axes([d,d,1-2*d,1-2*d])
sns.set_style("whitegrid")
tips_mx = tips.corr()
ax = sns.clustermap(tips_mx, figsize=(2,2))
ax.fig.tight_layout()
plt.savefig('seaborn_clustermap.svg')
plt.show()
plt.clf()
```

```python
# lmplot plot
# -----------------------------------------------------------------------------

plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           markers=["o", "x"])

plt.savefig('seaborn_lmplot.svg')
plt.show()
plt.clf()
```

```python
# lmplot plot
# -----------------------------------------------------------------------------
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.set_axis_labels("Total bill", "Tip")
g.add_legend()

plt.savefig('seaborn_facetgrid.svg')
plt.show()
plt.clf()
```
